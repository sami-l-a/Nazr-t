## Test recommendations

### Subtask:
Modify the existing `recommend_test` function or create a new function to easily test recommendations for different product names.


**Reasoning**:
Define a function to encapsulate the recommendation logic for reusability.



```python
def recommend_products(product_name, cosine_sim_matrix, df, n=10):
    """
    Recommends top N products based on cosine similarity.

    Args:
        product_name (str): The name of the product for which to find recommendations.
        cosine_sim_matrix (np.ndarray): The cosine similarity matrix.
        df (pd.DataFrame): The original DataFrame containing product details.
        n (int): The number of top recommendations to return (excluding the input product).

    Returns:
        pd.DataFrame: A DataFrame containing the top N recommended products with their details and scores.
    """
    # Handle potential multiple occurrences of the product name
    product_indices = df.index[df['name'] == product_name].tolist()

    if not product_indices:
        print(f"Product '{product_name}' not found in the dataset.")
        return pd.DataFrame(columns=['name', 'score', 'category', 'description'])

    # Get the average similarity scores if multiple indices exist, otherwise use the single index
    if len(product_indices) > 1:
        avg_scores = np.mean(cosine_sim_matrix[product_indices], axis=0)
        scores = list(enumerate(avg_scores))
    else:
        scores = list(enumerate(cosine_sim_matrix[product_indices[0]]))

    # Sort the scores in descending order and exclude the input product(s)
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    # Remove scores of the input product itself
    sorted_scores = [item for item in sorted_scores if item[0] not in product_indices]


    # Get the indices of the top N recommended products
    selected_indices = [i[0] for i in sorted_scores[:n]]
    selected_scores = [i[1] for i in sorted_scores[:n]]

    # Retrieve the details of the recommended products
    recommended_products_df = df.iloc[selected_indices].copy()
    recommended_products_df['score'] = selected_scores

    return recommended_products_df[['name', 'score', 'category', 'description']]
```

**Reasoning**:
Test the newly created `recommend_products` function with an example product name to ensure it works correctly and displays the recommended products.



```python
# Test the function with a product name
recommended_products = recommend_products('Wedding Candle', cosine_sim_mat, df, n=10)
display(recommended_products)
```
